```python
%pip install scipy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy as sp
from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/data/pinguinos.csv')
df =df.dropna()
df
```

```python
# Calcular el coheficiente de correlación
r, p =sp.stats.pearsonr(df['flipper_length_mm'],df['culmen_length_mm'])
r,p
```

```python
#Gráfico de regresión
sns.regplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    scatter_kws={
        's':10
    }
)
#Ahora vamos a colocar  sobre el gráfico, el valor que nos dio r
texto ='r='+str(r)[0:4]
plt.text(220, 35,texto,color='black')
```
